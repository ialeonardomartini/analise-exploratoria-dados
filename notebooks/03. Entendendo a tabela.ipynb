# Entendendo a tabela

## Objetivo: Conhecer mais a fundo o que são os dados da tabela.


```python
import os
import pandas as pd

repo_dir = os.path.abspath(os.path.join(os.getcwd(), ".."))
file_patch = os.path.join(repo_dir, "materiais", "survey_results_public.csv")
df = pd.read_csv(file_patch, index_col="ResponseId")
df
```

### Tipos de dados nas colunas
Para entender os tipos de dados, vamos usar `df.info()`:

```python
df.info()
```

Vemos que quase todas as são do tipo `object`. Geralmente, isto significa que representam texto. Logo mais nos aprofundaremos nisso.

Conseguimos ver que muitas colunas possuem valores nulos. Vamos explorar onde estes valores estão:

### Lidando com valores nulos

Podemos usar `df.isna()` para preencher o DataFrame com valores True/False, de acordo com o valor em cada célula ser nulo ou não:

```python
df.isna()
```

Como `True` é equivalente ao número `1` em Python, podemos simplesmente usar o método `df.sum()` para encontrar o número de valores nulos em cada coluna:

```python
df.isna().sum()
```

Podemos dividir esse valor pelo número de linhas para obter o valor percentual:

```python
df.isna().sum() / len(df)
```

E finalmente, podemos ordenar e adicionar um pouco de formatação ao output.

Fazemos isso definindo a nossa própria função de formatação de valores percentuais, e depois a entregamos para o método `apply()`:

```python
def format_percent(value: float) -> str:
    """Retorna o float formatado como valor percentual."""
    return f'{100 * value:.2f}%'


(df.isna().sum() / len(df)).sort_values(ascending=False).apply(format_percent)
```

Os nossos dados possuem muitos valores faltantes - em alguns casos, mais de 50% das linhas!

Neste curso, seguiremos com todos os dados. Mas na vida real, provavelmente gostaríamos de eliminar os valores nulos, seja removendo linhas e/ou colunas do nosso conjuno de dados, ou até mesmo substituindo os valores nulos por outros valores.

Para finalizar, vamos visualizar essa informação em um gráfico de barra simples:

```python
import matplotlib.pyplot as plt
import seaborn as sns

percent_data = (df.isna().sum() / len(df)).sort_values(ascending=False).reset_index()
percent_data.columns = ['column_name', 'percent_na']


fig, ax = plt.subplots(figsize=(18, 8))

sns.barplot(data=percent_data, ax=ax, x='column_name', y='percent_na')
fig.suptitle('% de dados faltantes em cada coluna')
plt.xticks(rotation=80)

plt.show()
```
